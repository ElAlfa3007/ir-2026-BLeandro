### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
path2 = 'C:\\Users\\leand\\Desktop\\ir2026\\ir-2026-BLeandro\\ir26a-main\\data\\gutenberg\\data'

txt_files = []
for root, _, files in os.walk(path2):
    for filename in files:
        if filename.lower().endswith('.txt'):
            txt_files.append(os.path.join(root, filename))

txt_files.sort()

corpus_path = os.path.join(path2, 'corpus_gutenberg_1000.txt')
with open(corpus_path, 'w', encoding='utf-8') as out_file:
    for txt_file in txt_files:
        with open(txt_file, 'r', encoding='utf-8', errors='ignore') as in_file:
            text = in_file.read().strip()
            if text:
                out_file.write(text.replace('\n', ' ') + '\n')

with open(corpus_path, encoding='utf-8') as f:
    documents = [line.strip() for line in f if line.strip()]

print(f'Corpus construido en: {corpus_path}')
print(f'Documentos en el corpus Gutenberg 1000: {len(documents)}')
print(f"palabras únicas en el corpus: {len(set(' '.join(documents).split()))}")
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
# Leer el corpus
with open(path2 + '\\corpus_gutenberg_1000.txt', encoding='utf-8') as f:
    documents2 = [line.strip() for line in f if line.strip()]

vectorizer = TfidfVectorizer(
    max_features=10000, 
    max_df=0.85,        
    min_df=3             
)

# Crear matriz TF-IDF (SPARSE)
tfidf_matrix = vectorizer.fit_transform(documents2)

print("Shape de la matriz:", tfidf_matrix.shape)

# Si quieres ver una parte como DataFrame (solo muestra pequeña)
sample_df = pd.DataFrame(
    tfidf_matrix[:5, :20].toarray(),  # solo 5 documentos y 20 palabras
    columns=vectorizer.get_feature_names_out()[:20]
)

sample_df.head()
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def buscar(query, modelo_vectorizador, matriz_tfidf, documentos, archivos, top_k=5):
    # 1. Vectorizar la consulta 
    query_vec = modelo_vectorizador.transform([query])
    
    # 2. Calcular similitud coseno contra TODOS los documentos
    similarities = cosine_similarity(query_vec, matriz_tfidf).flatten()
    
    # 3. Ordenar por relevancia (de mayor a menor)
    top_indices = np.argsort(similarities)[::-1][:top_k]
    
    # 4. Mostrar resultados
    resultados = []
    for idx in top_indices:
        resultados.append({
            "score": similarities[idx],
            "libro": os.path.basename(archivos[idx]), 
            "texto": documentos[idx][:50]  
        })
    
    return resultados
```

```python
res = buscar("Frankenstein", vectorizer, tfidf_matrix, documents2, txt_files, top_k=5)

for r in res:
    print(f"Libro: {r['libro']}")
    print(f"Score: {r['score']:.4f}")
    print(f"Texto: {r['texto']}...")
    print("-" * 60)
```
